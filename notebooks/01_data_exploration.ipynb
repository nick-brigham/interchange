```python
import pandas as pd
#import openpyxl

df = pd.read_excel('../data/raw/Interchange Training File.xlsx')
print(df.shape)
print(df.head())
```

```python
#missing values
print("Missing Values")
print(df.isnull().sum())

#duplications
print(f"\nDuplicate rows: {df.duplicated().sum()}")

#description length variation
df.columns = ['customer_description','part_number']
print(f"Shortest description: {df['customer_description'].str.len().min()}")
#reminder for self x.str > applies string logic
# x.len() gives length of string
# x.len().min() gives minimum vlaue of length from DF
print(f"\nLongest description: {df['customer_description'].str.len().max()}")
print(df[df['customer_description'].str.len() > 30])
```

```python
#cleaning script
df = df.dropna(subset=['part_number'])
print(f"After dropping NaN: {len(df)} rows")

df = df.drop_duplicates()
print(f"\nAfter dropping duplicates: {len(df)} rows")

#remove white space from cells
df['customer_description'] = df['customer_description'].str.strip()
df['part_number'] = df['part_number'].str.strip()

#remove any rows where either column string is empty
df = df[df['customer_description'] != '']
df = df[df['part_number'] != '']
print(f"\nAfter dropping empty cells: {len(df)} rows")

print(f"\nFinal clean rows: {len(df)} rows")
print(f"\nSample of clean data:")
df.head(10)
```

```python
from sklearn.model_selection import train_test_split

#split 80% train & 20% test
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

print(f"Training rows: {len(train_df)}")
print(f"Test rows: {len(test_df)}")

#save files
train_df.to_csv('../data/processed/train.csv', index=False)
test_df.to_csv('../data/processed/test.csv', index=False)

#save test input only, will compare to output later for validation.
test_df[['customer_description']].to_csv('../data/test/test_input.csv', index=False)
test_df.to_csv('../data/test/test_answers.csv', index=False)

print("\nFiles saved successfully")
```
